Create a Linear Regression model using Python/R to predict home prices using Boston Housing dataset (kaggle link: https://www.kaggle.com/datasets/altavish/boston-housing-dataset?resource=download)
The Boston housing dataset contains information about various houses in Boston through different parameters.There are 506 samples and 14 feature variables in this dataset.
The objectives is to predict the value of prices of the house using the given features

Defining the dependent and independent variables

```python
X = df.drop(['MEDV'], axis = 1)
Y = df['MEDV']
```

```python
X
```

```python
Y
```

Importing the dependecies for LinearRegression model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
```

Splitting the data into Train and Test

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20,random_state=42)
```

```python
print(X.shape, X_train.shape, X_train.shape)
```

Model fitting and implementation

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)     
```

```python
#Accuracy on training data
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

Checking the model score

```python
model.score(X_test,Y_test)
```
